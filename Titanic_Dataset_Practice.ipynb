```python
df['Age'].isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True) #insert average in null value column // Its called MEAN()
```

```python
df['Age'].isnull().sum()
```

```python
#binary text column replace with 0-1 // its called MAP

df['Sex']=df['Sex'].map({'male':1, 'female':0})
```

```python
df.head()
```

```python
#adding new column or modify column (apply lambda function)

df['Last Name']=df['Name'].apply(lambda x: x.split(', ')[0])
df['First Name']= df['Name'].apply( lambda x: x.split(', ')[1])

```
